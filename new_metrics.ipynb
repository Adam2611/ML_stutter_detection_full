```python
def count_limits(upper_limit):
    stutter_limit_count = 0
    nonstutter_limit_count = 0
    times = 4
    for idx in range(len(train_x)):
        counter = 0
        for num in train_x[idx]:
            if num[0] > upper_limit:
                counter+=1
                if counter>times and train_y[idx] == 0:
                    nonstutter_limit_count +=1
                    break
                elif counter>times and train_y[idx] == 1:
                    stutter_limit_count+=1
                    break
    
    for idx2 in range(len(test_x)):
        counter2 = 0
        for num2 in test_x[idx2]:
            if num2[0] > upper_limit:
                counter2+=1
                if counter2>times and test_y[idx2] == 0:
                    nonstutter_limit_count +=1
                    break
                elif counter2>times and test_y[idx2] == 1:
                    stutter_limit_count+=1
                    break

    # print(stutter_limit_count)
    # print(nonstutter_limit_count)
    return stutter_limit_count, nonstutter_limit_count
```

```python
stutter_limit_list = []
nonstutter_limit_list = []
for stutter_ms in range(1,150):
    limits = count_limits(stutter_ms)
    stutter_limit_list.append(limits[0])
    nonstutter_limit_list.append(limits[1])
```

```python
stutter_limit_list = stutter_limit_list[0:130]
nonstutter_limit_list = nonstutter_limit_list[0:130]
plt.plot(stutter_limit_list, color='blue', marker='o',mfc='green' ) #plot the data
plt.plot(nonstutter_limit_list, color='red', marker='o',mfc='orange' ) #plot the data
plt.xticks(range(1, 100, 1)) #set the tick frequency on x-axis

plt.ylabel('Count') #set the label for y axis
plt.xlabel("MS Limit") #set the label for x-axis
plt.title("Counts of Limit vs Absolute Limit MS") #set the title of the graph

plt.show() #display the graph
plt.axis('off')

spec = plt.imshow

plt.savefig('spec',bbox_inches='tight',transparent=True, pad_inches=0)
```
